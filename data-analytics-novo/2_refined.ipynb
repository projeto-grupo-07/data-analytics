# 2. Refinamento de Dados — Trusted → Refined

Parte do CSV trusted (dataset completo tratado) e aplica os filtros de negócio:

- Filtra apenas calçados (NCM começa com **64**)
- Seleciona e renomeia as colunas relevantes para análise
- Salva o resultado em `refined/`

## 1. Importação das bibliotecas

```python
import pandas as pd
```

## 2. Carregamento do CSV trusted

```python
df = pd.read_csv('trusted/estatisticas_di_2026_1_trusted.csv', dtype={'ncm_str': str})

print(f'Shape trusted: {df.shape}')
df.head()
```

## 3. Filtro de calçados (NCM 64)

```python
df_shoes = df[df['ncm_str'].str.startswith('64')].copy()

print(f'Registros de calçados: {len(df_shoes)} (de {len(df)} totais)')
```

## 4. Seleção e renomeação de colunas

```python
df_refined = df_shoes[[
    'ncm', 'pais_origem', 'qtd_ume_soma_truncado', 'vmle_soma_dolar_truncado',
    'fob_ume_media_dolar_truncado', 'frete_media_dolar_truncado', 'seguro_media_dolar_truncado'
]].copy()

df_refined.columns = [
    'ncm', 'pais_origem', 'volume_total', 'valor_total',
    'preco_medio', 'frete_medio', 'seguro_medio'
]

print(f'Shape refined: {df_refined.shape}')
df_refined.head()
```

## 5. Salva em refined/

```python
df_refined.to_csv('refined/estatisticas_di_2025_10-12_refined.csv', index=False)
print(f'✔ CSV refined salvo — {len(df_refined)} registros')
```
