```python
import pandas as pd

# 1. Carrega os dados de Declaração de Importação de Outubro a Dezembro de 2025
df = pd.read_csv('estatisticas_di_2025_10-12.csv')

# 2. Filtra os dados apenas para mostrar os calçados (segundo Nomenclatura Comum do Mercosul (NCM), os calçados começam com o codigo 64 )
df['ncm_str'] = df['ncm'].astype(str).str.zfill(8)
only_shoes_filter_df = df[df['ncm_str'].str.startswith('64')]

# 3. Filtra as colunas mais relevantes
only_shoes_filter_best_columns_df = only_shoes_filter_df[['ncm', 'pais_origem', 'qtd_ume_soma_truncado', 'vmle_soma_dolar_truncado', 'fob_ume_media_dolar_truncado', 'frete_media_dolar_truncado', 'seguro_media_dolar_truncado']].copy()

# 4. Renomeia as colunas para facilitar
only_shoes_filter_best_columns_df.columns = ['ncm', 'pais_origem', 'volume_total', 'valor_total', 'preco_medio', 'frete_medio', 'seguro_medio']

# 5. Print da primeira parte dos dados para visualização
only_shoes_filter_best_columns_df.head()

# 6. Carregamento dos dados trusted prontos para analise.
only_shoes_filter_best_columns_df.to_csv('estatisticas_di_2025_10-12_trusted.csv')
```
