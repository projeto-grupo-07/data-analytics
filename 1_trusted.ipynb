# 1. Tratamento de Dados — Raw → Trusted

Carrega o CSV bruto da Receita Federal e aplica tratamentos gerais **sem filtrar** por categoria.

- Normaliza nomes de colunas
- Garante tipos corretos
- Adiciona coluna `ncm_str` (NCM zerado à esquerda)
- Salva o resultado em `trusted/`

## 1. Importação das bibliotecas

```python
import pandas as pd
```

## 2. Carregamento do CSV bruto

```python
df = pd.read_csv('raw/estatisticas_di_2025_10-12.csv')

print(f'Shape original: {df.shape}')
df.head()
```

## 3. Tratamento geral

```python
# Normaliza nomes de colunas
df.columns = df.columns.str.lower().str.strip()

# Garante tipo string no NCM e adiciona versão zerada à esquerda
df['ncm'] = df['ncm'].astype(str).str.strip()
df['ncm_str'] = df['ncm'].str.zfill(8)

# Normaliza país de origem
df['pais_origem'] = df['pais_origem'].astype(str).str.strip().str.upper()

print(f'Shape tratado: {df.shape}')
print(f'Colunas: {list(df.columns)}')
df.head()
```

## 4. Salva em trusted/

```python
df.to_csv('trusted/estatisticas_di_2025_10-12_trusted.csv', index=False)
print(f'✔ CSV trusted salvo — {len(df)} registros')
```
